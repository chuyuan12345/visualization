```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### bar chart

```python
# Data for the bar chart
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
values = [25, 30, 22, 35]

# Create a bar chart using seaborn
sns.barplot(x=categories, y=values)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example using Seaborn')

# Display the chart
plt.show()
```

### pie chart

```python
# Data for the pie chart
labels = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
sizes = [25, 30, 22, 35]

# Create a pie chart using seaborn
plt.figure(figsize=(6, 6))  # Set figure size
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart Example using Seaborn')

# Display the pie chart
plt.show()
```

### line chart

```python
# Data for the line chart
x = [1, 2, 3, 4, 5]
y1 = [25, 30, 22, 35, 40]
y2 = [40, 35, 22, 30, 25]

# Create a line chart using seaborn
plt.figure(figsize=(8, 6))  # Set figure size
sns.lineplot(x=x, y=y1, label='Line 1')
sns.lineplot(x=x, y=y2, label='Line 2')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Line Chart Example using Seaborn')
plt.legend()

# Display the line chart
plt.show()
```

### scatter plot

```python
# Data for the scatter plot
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]

# Create a scatter plot using seaborn
sns.scatterplot(x=x, y=y)
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Scatter Plot Example using Seaborn')

# Display the scatter plot
plt.show()
```

### bubble chart

```python
# Data for the bubble chart
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]
sizes = [30, 160, 190, 320, 500]  # Sizes for the "bubbles"

# Create a scatter plot with varying point sizes using seaborn
sns.scatterplot(x=x, y=y, s=sizes)
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Bubble Chart Example using Seaborn')

# Display the bubble chart
plt.show()
```
