```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# create sample data
data = {'Category': ['A', 'B', 'C', 'D'],
        'Count': [10, 20, 30, 40],
        'Color': ['red', 'green', 'blue', 'orange']}
df = pd.DataFrame(data)
```

```python
# create bar chart with labels and color coding
sns.barplot(x='Category', y='Count', data=df, palette=df['Color'])

# add labels to bars
for i, count in enumerate(df['Count']):
    plt.text(i, count + 1, str(count), ha='center')

# add axis labels and title
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Bar Chart with Labels and Color Coding')

# show the plot
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# create sample data
data = {'Category': ['A', 'B', 'C', 'D'],
        'Count': [10, 20, 30, 40],
        'Color': ['red', 'green', 'blue', 'orange']}
df = pd.DataFrame(data)

# create bar chart with custom colors and legend
sns.barplot(x='Category', y='Count', data=df, palette=df['Color'])

# set legend
plt.legend(title='Category Colors', labels=df['Category'])
# legend.get_texts()[0].set_color("red")
# legend.get_texts()[1].set_color("green")
# legend.get_texts()[2].set_color("blue")
# legend.get_texts()[3].set_color("orange")

# add axis labels and title
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Bar Chart with Legend')

# show the plot
plt.show()
```

### legend (to be further explored)

```python
import matplotlib.pyplot as plt

# create sample data
x = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
y2 = [20, 30, 10, 50]

# create line plots with different colors
plt.plot(x, y1, color='red', label='Line 1')
plt.plot(x, y2, color='blue', label='Line 2')

# set legend with custom colors and labels
legend = plt.legend(loc='upper left', facecolor='gray')
legend.get_frame().set_alpha(0.5)
legend.get_texts()[0].set_color("red")
legend.get_texts()[1].set_color("blue")
legend.get_texts()[0].set_text("First Line")
legend.get_texts()[1].set_text("Second Line")

# add axis labels and title
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Chart with Custom Legend')

# show the plot
plt.show()
```
